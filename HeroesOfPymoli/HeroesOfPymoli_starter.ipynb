
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_df = df.groupby ('Gender')
count = gender_purchase_df['Gender'].count()

gender_avg_price = gender_purchase_df['Price'].mean()
gender_pur_sum = gender_purchase_df['Price'].sum()
avgprice_usergender = gender_pur_sum/count

gender_pur_summary_df = pd.DataFrame(count)
gender_pur_summary_df['Average Price by Gender']=gender_avg_price
gender_pur_summary_df['Total Purchases by Gender']=gender_pur_sum
gender_pur_summary_df['Average Total Purchase Price by Gender']=avgprice_usergender

gender_pur_summary_df = gender_pur_summary_df.rename(columns = {'Gender': 'Purchase Count',})

gender_pur_summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,50]
groups = ['<10', '10-14', '15-19','20-24', '25-29','30-34','35-39','40+']

df['Age Group']=pd.cut(df['Age'],bins, labels=groups)

age_df = df.groupby(['Age Group'])

age_purchase = age_df['Price'].count()
average_price_by_age = age_df['Price'].mean()
total_spent = age_df['Price'].sum()
avg_by_person_by_age = total_spent / age_purchase

summary_age_df = pd.DataFrame(age_purchase)
summary_age_df ['Average Purchase by Age'] = average_price_by_age
summary_age_df ['Total Spent by Age'] = total_spent
summary_age_df['Average Spent Per Person by Age'] = avg_by_person_by_age

summary_age_df = summary_age_df.rename(columns = {'Price': 'Purchase Count'})

summary_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = df.groupby(['SN'])

purchase_per_user = top_spenders['Price'].count()
average_user_purchase = top_spenders['Price'].mean()
total_user_purchase = top_spenders['Price'].sum()

top_spenders_df = pd.DataFrame(purchase_per_user)
top_spenders_df ['Average Purchase Price by SN'] = average_user_purchase
top_spenders_df['Total Spent Per User'] = total_user_purchase

top_spenders_df = top_spenders_df.rename(columns= {'Price': 'Count of Purchases'})

#sorting highest to lowest 

top_spenders_df = top_spenders_df.sort_values(by = ['Total Spent Per User'], ascending=False)

top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items = df[['Item ID', 'Item Name', 'Price']]

popular_items = items.groupby(['Item ID', 'Item Name'])

purchase_count = popular_items['Price'].count()

purchase_value = popular_items['Price'].sum()

item_price = purchase_value / purchase_count

popular_summary = pd.DataFrame({'Purchase Count': purchase_count, 'Item Price': item_price, 'Total Purchase Value':purchase_value})

popular_summary = popular_summary.sort_values(['Purchase Count'], ascending=False).head()

popular_summary
```
